```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report

# Télécharger les données nécessaires pour le traitement
nltk.download('popular')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
nltk.download('omw-1.4')
```

```python
import spacy
```

```python
stopwords_english = set(stopwords.words('english'))
nlp = spacy.load('en_core_web_sm')
```

```python
#lemmatizer
def clean(text):
  sent_token = nlp(text.lower())
  tokens_clean_lemmatizer = [token.lemma_ for token in sent_token if not token.is_punct and not token.is_stop]
  return " ".join(tokens_clean_lemmatizer)
clean("You are better when I am well.")
```

```python
df['clean'] = (df['text']).apply(clean)
```

```python
df.head()
```

```python
X = df['clean']
y = df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=32)
```

```python
vectorize = CountVectorizer()
X_train_countV = vectorize.fit_transform(X_train)
X_test_countV = vectorize.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_logreg_count = LogisticRegression(random_state=32)
model_logreg_count.fit(X_train_countV, y_train)

# Prédictions
y_pred_count_logreg = model_logreg_count.predict(X_test_countV)

# Évaluation
print("Scores avec Logistic Regression et CountVectorizer :")
print("Accuracy:", accuracy_score(y_test, y_pred_count_logreg))
print(classification_report(y_test, y_pred_count_logreg))
```

```python
# Initialisation du TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
model_logreg = LogisticRegression(random_state=32,)
model_logreg.fit(X_train_tfidf, y_train)

y_pred_logreg = model_logreg.predict(X_test_tfidf)

print("Scores avec Logistic Regression et TfidfVectorizer :")
print("Accuracy:", accuracy_score(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))
```
